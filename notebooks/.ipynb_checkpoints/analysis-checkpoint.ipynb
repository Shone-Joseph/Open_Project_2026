```python
import requests
import pandas as pd

API_KEY = "layounwrepa"
BASE_URL = "https://api.europeana.eu/record/v2/search.json"

all_items = []

for page in range(1, 5):  # 4 pages × 100 rows = ~400
    params = {
        "wskey": API_KEY,
        "query": "India",
        "rows": 100,
        "page": page
    }

    response = requests.get(BASE_URL, params=params)
    data = response.json()

    for item in data.get("items", []):
        all_items.append({
            "title": item.get("title", [None])[0] if item.get("title") else None,
            "creator": item.get("dcCreator", [None])[0] if item.get("dcCreator") else None,
            "description": item.get("dcDescription", [None])[0] if item.get("dcDescription") else None,
            "subject": item.get("dcSubject", [None])[0] if item.get("dcSubject") else None,
            "dataProvider": item.get("dataProvider", [None])[0] if item.get("dataProvider") else None,
            "country": item.get("country"),
            "type": item.get("type"),
            "language": item.get("language", [None])[0] if item.get("language") else None
        })

df = pd.DataFrame(all_items)

print(len(df))
df.head()
```

```python
import requests
import pandas as pd

API_KEY = "layounwrepa"
BASE_URL = "https://api.europeana.eu/record/v2/search.json"

query_groups = {
    "core": ["India", "Indian", "Indien"],
    "historical": ["British India", "colonial India", "Mughal"],
    "heritage": ["Indian art", "South Asia", "Indian textiles"]
}

all_items = []
MAX_ROWS = 500

for group, queries in query_groups.items():
    for q in queries:
        page = 1

        while len(all_items) < MAX_ROWS:
            params = {
                "wskey": API_KEY,
                "query": q,
                "rows": 100,
                "page": page
            }

            response = requests.get(BASE_URL, params=params)
            data = response.json()

            items = data.get("items", [])

            if not items:
                break

            for item in items:
                all_items.append({
                    "query_group": group,
                    "search_term": q,
                    "title": item.get("title", [None])[0] if item.get("title") else None,
                    "description": item.get("dcDescription", [None])[0] if item.get("dcDescription") else None,
                    "subject": item.get("dcSubject", [None])[0] if item.get("dcSubject") else None,
                    "dataProvider": item.get("dataProvider", [None])[0] if item.get("dataProvider") else None,
                    "type": item.get("type")
                })

                if len(all_items) >= MAX_ROWS:
                    break

            page += 1

        if len(all_items) >= MAX_ROWS:
            break

df = pd.DataFrame(all_items)

print("Final dataset size:", len(df))
df.head()
```

```python
df.to_csv("data/europeana_india_dataset.csv", index=False)
```

```python
import requests
import pandas as pd

API_KEY = "layounwrepa"
BASE_URL = "https://api.europeana.eu/record/v2/search.json"

query_groups = {
    "core": ["India", "Indian", "Indien"],
    "historical": ["British India", "colonial India", "Mughal"],
    "heritage": ["Indian art", "South Asia", "Indian textiles"]
}

group_limits = {
    "core": 200,
    "historical": 150,
    "heritage": 150
}

all_items = []

for group, queries in query_groups.items():
    group_count = 0

    for q in queries:
        page = 1

        while group_count < group_limits[group]:

            params = {
                "wskey": API_KEY,
                "query": q,
                "rows": 100,
                "page": page
            }

            response = requests.get(BASE_URL, params=params)
            data = response.json()

            items = data.get("items", [])
            if not items:
                break

            for item in items:
                if group_count >= group_limits[group]:
                    break

                all_items.append({
                    "query_group": group,
                    "search_term": q,
                    "title": item.get("title", [None])[0] if item.get("title") else None,
                    "description": item.get("dcDescription", [None])[0] if item.get("dcDescription") else None,
                    "subject": item.get("dcSubject", [None])[0] if item.get("dcSubject") else None,
                    "dataProvider": item.get("dataProvider", [None])[0] if item.get("dataProvider") else None,
                    "type": item.get("type")
                })

                group_count += 1

            page += 1

df = pd.DataFrame(all_items)

print(df["query_group"].value_counts())
print(len(df))
```

```python
df.to_csv("data/europeana_india_dataset.csv", index=False)
```

```python
import requests
import pandas as pd
import os

# -----------------------------
# CONFIGURATION
# -----------------------------

API_KEY = "layounwrepa"
BASE_URL = "https://api.europeana.eu/record/v2/search.json"

query_groups = {
    "geographic": [
        "India",
        "Indian",
        "Indien",
        "Indische India"
    ],

    "colonial": [
        "British India",
        "colonial India",
        "British Raj",
        "East India Company",
        "colonial collection India"
    ],

    "cultural": [
        "Indian art",
        "Indian museum collection",
        "South Asia collection",
        "ethnography India",
        "India exhibition"
    ],

    "material": [
        "Indian textile",
        "Indian manuscript",
        "Indian photograph",
        "Indian map",
        "Indian print"
    ]
}

group_limits = {
    "geographic": 375,
    "colonial": 375,
    "cultural": 375,
    "material": 375
}

# -----------------------------
# STORAGE
# -----------------------------

all_items = []

# -----------------------------
# DATA COLLECTION LOOP
# -----------------------------

for group, queries in query_groups.items():
    group_count = 0

    print(f"Collecting group: {group}")

    for q in queries:
        page = 1

        while group_count < group_limits[group]:

            params = {
                "wskey": API_KEY,
                "query": q,
                "rows": 100,
                "page": page
            }

            response = requests.get(BASE_URL, params=params)

            if response.status_code != 200:
                print("API error:", response.status_code)
                break

            data = response.json()
            items = data.get("items", [])

            if not items:
                break

            for item in items:

                if group_count >= group_limits[group]:
                    break

                all_items.append({
                    "query_group": group,
                    "search_term": q,
                    "title": item.get("title", [None])[0] if item.get("title") else None,
                    "description": item.get("dcDescription", [None])[0] if item.get("dcDescription") else None,
                    "subject": item.get("dcSubject", [None])[0] if item.get("dcSubject") else None,
                    "dataProvider": item.get("dataProvider", [None])[0] if item.get("dataProvider") else None,
                    "type": item.get("type"),
                    "country": item.get("country")
                })

                group_count += 1

            page += 1

# -----------------------------
# DATAFRAME CREATION
# -----------------------------

df = pd.DataFrame(all_items)

print("\nFinal distribution:")
print(df["query_group"].value_counts())
print("\nTotal rows:", len(df))

# -----------------------------
# SAVE DATA
# -----------------------------

os.makedirs("data", exist_ok=True)

df.to_csv("data/europeana_india_dataset_1500.csv", index=False)

print("\nDataset saved successfully.")
```
